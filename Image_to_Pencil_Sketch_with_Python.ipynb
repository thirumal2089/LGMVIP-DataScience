Task 4 : Image to Pencil Sketch with Python

Dataset Link : https://i.pinimg.com/originals/93/98/75/9398754385c6708cb744a17644741456.jpg

**Importing Libraries**

```python
import cv2
from google.colab.patches import cv2_imshow as cvm
import matplotlib.pyplot as plt
from skimage import io
print("Libraries imported successfully")
```

```python
link="https://i.pinimg.com/originals/93/98/75/9398754385c6708cb744a17644741456.jpg"
```

**Loading an image of Swami Vivekananda**

```python
image=io.imread(link)
```

```python
image
```

```python
cvm(image)
```

**Reducing the size of the image**

```python
image = cv2.resize(image,(500,450))
cvm(image)
```

**Converting into Gray Scale**

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_image
```

```python
cvm(gray_image)
```

**Inverting the image**

```python
inverted_image = cv2.bitwise_not(gray_image)
inverted_image
```

```python
cvm(inverted_image)
```

**Smoothing the image**

```python
image_smoothing = cv2.GaussianBlur(inverted_image, (21, 21), sigmaX=0, sigmaY=0)
image_smoothing
```

```python
cvm(image_smoothing)
```

**Pencil Sketch of the image**

```python
def pencil_sketch(x, y):
  return cv2.divide(x, 255 - y, scale=256)
```

```python
pencil_sketch = pencil_sketch(gray_image, image_smoothing)
cvm(pencil_sketch)
```

**Conclusion** : Thus we can convert the image into pencil sketch using Python.
